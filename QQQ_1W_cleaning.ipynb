```python
import pandas as pd
```

```python
df  = pd.read_csv('data/input/BATS_QQQ, 1W.csv')
```

## DF overview

```python
df.info()
```

## General DataFrame Cleaning

```python
df.tail(2)
```

```python
df['date'] = pd.to_datetime(df['time'])
```

```python
df.tail(2)
```

```python
df.head(2)
```

```python
df.drop(columns=['US07Y, TVC: Open',
                 'US07Y, TVC: High', 
                 'US07Y, TVC: Low',
                 'US07Y, TVC: Close',
                'time'], inplace=True)
```

## DF Slicing

```python
close = df[['date',
            'close',
           'US05Y, TVC: Close',
           'US10Y, TVC: Close',
           'US30Y, TVC: Close',
           'Volume']]
```

```python
open_ = df[['date',
            'close',
           'US05Y, TVC: Open',
           'US10Y, TVC: Open',
           'US30Y, TVC: Open',
           'Volume']]
```

```python
high = df[['date',
            'close',
           'US05Y, TVC: High',
           'US10Y, TVC: High',
           'US30Y, TVC: High',
           'Volume']]
```

```python
low = df[['date',
            'close',
           'US05Y, TVC: Low',
           'US10Y, TVC: Low',
           'US30Y, TVC: Low',
           'Volume']]
```

### DF's to CSV's

```python
close.to_csv('data/processed/qqq_close')
open_.to_csv('data/processed/qqq_open')
high.to_csv('data/processed/qqq_high')
low.to_csv('data/processed/qqq_low')
```
